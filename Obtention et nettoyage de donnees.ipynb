```python
import pandas as pd
import numpy as np
```

This code generate the fictive data for the project

```python
# Configuration des paramètres
num_clients = 1000  # Nombre de clients à générer

# Génération de données fictives
np.random.seed(42)  # Pour la reproductibilité

client_ids = np.arange(1, num_clients + 1)
revenus = np.random.randint(20000, 120000, size=num_clients)  # Revenu annuel entre 20k et 120k
transactions_montant = np.random.uniform(100, 5000, size=num_clients)  # Montant moyen des transactions
frequence_achats = np.random.randint(1, 30, size=num_clients)  # Fréquence d'achats par mois
historique_credit = np.random.choice(['Bon', 'Moyen', 'Mauvais'], size=num_clients, p=[0.6, 0.3, 0.1])  # Historique de crédit
taux_endettement = np.random.uniform(0.1, 0.5, size=num_clients)  # Taux d'endettement entre 10% et 50%
age_client = np.random.randint(18, 70, size=num_clients)  # Âge entre 18 et 70 ans

# Nouvelles caractéristiques
etat_civil = np.random.choice(['Célibataire', 'Marié', 'Divorcé'], size=num_clients)
nombre_enfants = np.random.randint(0, 6, size=num_clients)  # Entre 0 et 5 enfants
type_emploi = np.random.choice(['Salarié', 'Indépendant', 'Chômeur'], size=num_clients)
duree_emploi = np.random.randint(0, 30, size=num_clients)  # Durée en années dans l'emploi actuel
proprietaire_maison = np.random.choice(['Oui', 'Non'], size=num_clients)
score_credit = np.random.randint(300, 850, size=num_clients)  # Score de crédit entre 300 et 850
retards_paiement = np.random.poisson(0.5, size=num_clients)  # Nombre moyen de retards

# Création du DataFrame
data = {
    'Client_ID': client_ids,
    'Revenu_Annuel': revenus,
    'Montant_Transactions': transactions_montant,
    'Frequence_Achats': frequence_achats,
    'Historique_Credit': historique_credit,
    'Taux_Endettement': taux_endettement,
    'Age_Client': age_client,
    'Etat_Civil': etat_civil,
    'Nombre_Enfants': nombre_enfants,
    'Type_Emploi': type_emploi,
    'Duree_Emploi': duree_emploi,
    'Proprietaire_Maison': proprietaire_maison,
    'Score_Credit': score_credit,
    'Retards_Paiement': retards_paiement
}

df_enrichi = pd.DataFrame(data, columns=['Client_ID', 'Revenu_Annuel', 'Montant_Transactions', 'Frequence_Achats', 'Historique_Credit', 'Taux_Endettement', 'Age_Client', 'Etat_Civil', 'Nombre_Enfants', 'Type_Emploi', 'Duree_Emploi', 'Proprietaire_Maison', 'Score_Credit', 'Retards_Paiement'])

# Exportation vers un fichier CSV (optionnel)
df_enrichi.to_csv("/content/drive/MyDrive/Projet Systeme d'evaluation de risque de credit/donnees_fictives_risque_credit_enrichies.csv", index=False)

# Affichage des premières lignes du DataFrame enrichi
df_enrichi
```
